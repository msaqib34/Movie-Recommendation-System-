```python
def genre(name):
    l=[]
    for i in eval(name):
        l.append(i['name'])
    return l
```

```python
def cast(name):
    l=[]
    count=0
    for i in eval(name):
        if count!=3:
            l.append(i['name'])
            count+=1
    return l
```

```python
def crew(name):
    l=[]
    for i in eval(name):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
def keywords(name):
    l=[]
    count=0
    for i in eval(name):
        if count!=5:
            l.append(i['name'])
            count+=1
    return l
```

```python
movies_data['keywords']=movies_data['keywords'].apply(keywords)
```

```python
movies_data=movies_data.dropna()
```

```python
movies_data['overview']=movies_data['overview'].apply(lambda x:x.split())
```

```python
movies_data['genres']=movies_data['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies_data['cast']=movies_data['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies_data['crew']=movies_data['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies_data['overview']=movies_data['overview'].apply(lambda x: [i.replace(" ","") for i in x])
movies_data['keywords']=movies_data['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies_data.head(1)
```

```python
movies_data['tags']=movies_data['overview']+movies_data['genres']+movies_data['cast']+movies_data['keywords']+movies_data['crew']
```

```python
movies_data=movies_data[['id','title','imdb_id','tags']]
```

```python
movies_data.drop(columns='id', inplace=True)
```

```python
movies_data
```

```python
movies_data['tags']=movies_data['tags'].apply(lambda x: " ".join(x))
```

```python
movies_data
```

```python
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features= 5000 , stop_words='english')
```

```python
vector=cv.fit_transform(movies_data['tags']).toarray()
```

```python
vector
```

```python
# importing cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
def recommend(movie):
  # Movie index
  movie_index= movies_data[movies_data['title']==movie].index[0]
  # Checking similarity between movies
  best_match= similarity[movie_index]
  # Sorting the movies
  movie_list=sorted(list(enumerate(best_match)),reverse=True, key=lambda x:x[1])[1:6]

  for i in movie_list:
    # Printing the name of movies
    print(movies_data.iloc[i[0]].title)
```

```python
recommend('Jumanji')
```
